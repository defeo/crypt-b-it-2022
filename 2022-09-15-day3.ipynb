### Determine the endomorphism ring of the curve below

Hint: start by computing the discriminant of Frobenius

```python
p = 13172209
E = EllipticCurve(GF(p), j=3624559)
E
```

We compute the discriminant and the conductor of $ℤ[π]$

```python
O = E.frobenius_order()
O
```

```python
O.discriminant().factor()
```

```python
O.index_in(O.number_field().maximal_order()).factor()
```

Our $\mathrm{End}(E)$ is contained between $ℤ[π]$ and the maximal order of $ℚ(\sqrt{-59})$. For each of the primes in the conductor of $ℤ[π]$, we determine how many times it divides the conductor of $\mathrm{End}(E)$ by looking at the level of $E$ in its $ℓ$-isogeny graph.

We start with $ℓ = 5$. To plot the whole graph in a readable way, we need to play a bit with the options to the `.plot()` function. We find that $E$ (the red vertex) is at the bottom, and thus $5|[\mathcal{O}_{-59}:\mathrm{End}(E)]$.

```python
G = E.isogeny_ell_graph(5, label_by_j=True, directed=False)
```

```python
G.plot(vertex_labels=False,
       layout="spring",
       iterations=1000,
       vertex_colors={ "red" : [str(E.j_invariant())] },
       figsize=(10,10),
      )
```

We could have come to the same conclusion by only looking at the neighbours of $E$. It has only one neighbour in the $5$-isogeny graph, thus it must be at the bottom.

```python
E.isogenies_prime_degree(5)
```

Because $7$ divides the conductor of $π$ only once, the $7$-volcano also has height $1$. The same reasoning as before tells us that $E$ is at the top of the volcano, and thus $7 \nmid [\mathcal{O}_{-59}:\mathrm{End}(E)]$.

```python
E.isogenies_prime_degree(7)
```

For $ℓ=3$ the volcano has $4$ levels. Exploring the neighbours tells us that $E$ is not at the bottom, but it is not sufficient to decide at which level it is.

```python
E.isogenies_prime_degree(3)
```

Going one step further shows us that three of the four neighbours of $E$ only have one neighbour, thus they must be on the floor, hence $E$ is one level above the floor. Thus $9 | [\mathcal{O}_{-59}:\mathrm{End}(E)]$. We conclude that $\mathrm{End}(E)$ has conductor $45$.

```python
[len(i.codomain().isogenies_prime_degree(3)) for i in E.isogenies_prime_degree(3)]
```

#### 3. Starting from $E$, do a random walk in the $(5,7,11,13,17)$-isogeny graph

We define a function that takes in input the number of times each degree $ℓ ∈ [5,7,11,13,17]$ is to be taken, and then performs that many steps. As implemented here, the walk may backtrack, i.e., follow an isogeny with its dual.

```python
def random_walk(E, exponents):
    from random import choice
    for ell, exp in zip([5, 7, 11, 13, 17], exponents):
        for i in range(exp):
            isos = E.isogenies_prime_degree(ell)
            print(E.j_invariant(), end="->")
            E = choice(isos).codomain()
    return E
```

#### 4. Let $π$ be the Frobenius of $E$. Let $\mathfrak{a} = (π-1,5)$ be the ideal of $\mathrm{End}(E)$ generated by $π-1$ and $5$. Determine $E[\mathfrak{a}]$.

By definition, $E[\mathfrak{a}] = E[5] ∩ E[π-1]$. The first group is the $5$-torsion subgroup (over the algebraic closure), the second is the subgroup of $𝔽_p$-rational points. So we just need to find the $𝔽_p$-rational points of order $5$.

A first method is to invert the multiplication-by-5 polynomials, which can be done as follows:

```python
E(0).division_points(5)
```

A second method is by taking a generator of $E(𝔽_p)$, and multiplying by the cofactor of $5$

```python
P, = E.gens()
g = (E.order() // 5) * P
[g*i for i in range(5)]
```

#### 5. Find a path in the $(5,7,11,13,17)$-isogeny graph between $E$ and the curve of $j$-invariant $197819$.

The `random_walk` function above is very bad for this task, as it would do lots of useless backtracking.

To mimic a real-world attack, we should implement non-backtracking random walks, and then implement a collision based algorithm. But here we can rely on the fact that we have a toy instance, and just walk in the rims of the respective volcanos until we reach $197819$. Even for this, we need to implement non backtracking walks on the craters.

To define a direction on the craters, we look at how Frobenius acts modulo each $ℓ$.

```python
[E.frobenius_polynomial().change_ring(GF(ell)).roots() for ell in (5,7,11,13,17)]
```

Aha! The polynomial splits as $(π-1)(π-4)$ modulo each of the primes. Thus we can proceed as before: define the ideal $\mathfrak{a} = (π-1,ℓ)$, then $E[\mathfrak{a}]$ is the subgroup of $𝔽_p$-rational points of order $ℓ$. These groups define isogeny walks that never backtrack.

```python
def walk_crater(E, ell, steps=1):
    for i in range(steps):
        print(E.j_invariant(), end="->")
        gen = E(0).division_points(ell)[-1]
        E = E.isogeny(gen).codomain()
    return E
```

```python
walk_crater(E, 5, 10)
```

```python
def find_dist(E, F, ell):
    i = 0
    while E.j_invariant() != F.j_invariant():
        E = walk_crater(E, ell)
        i += 1
    return i
```

```python
[find_dist(E, EllipticCurve(GF(p), j=197819), ell) for ell in (5,7,11,13,17)]
```

Factoring the Frobenius polynomial, we see that there exists precisely one cyclic subgroup of $E[19]$ whose points are defined over $𝔽_{p^2}$.

#### 7. Find a path in the $19$-isogeny graph betwee $E$ and $j=197819$. What equality do you deduce in the class group of $\mathrm{End}(E)$?

We proceed as before, but we try yet another method to compute the isogeny. Instead of finding a point generating the kernel, we find the polynomial that vanishes on it. Because we know that the $x$-coordinates of $E[(π+1,19)]$ are defined over $𝔽_p$, this is equal to the GCD of the $19$th division polynomial with $X^p - X$.

```python
def walk_19(E, steps=1):
    R.<x> = GF(p)[]
    for i in range(steps):
        ker = E.division_polynomial(19).gcd(x^p - x)
        print(E.j_invariant(), end="->")
        E = E.isogeny(ker).codomain()
    return E
```

$$(π-1,5)^{258} = (π-1,7)^{60} = (π-1,11)^{16} = (π-1,13)^{71} = (π-1,17)^{348} = (π+1,19)$$

in the class group.

#### 8. Implement a key exchange using the CM graph. Check that Alice and Bob compute the same shared key.

A function to generate a random tuple of $5$ exponents in $[0...3]$

```python
def secret_gen():
    from random import choices
    return choices(range(4), k=5)
```

A function to compute the group action by the ideal $\prod (π-1, ℓ)^{e_ℓ}$, with $e_ℓ$ a list of exponents given in input.

```python
def group_action(E, exp):
    for (e, ell) in zip(exp, (5,7,11,13,17)):
        E = walk_crater(E, ell, e)
    return E
```

```python
# Alice
skA = secret_gen()
pkA = group_action(E, skA)

# Bob
skB = secret_gen()
pkB = group_action(E, skB)

# Shared secret
group_action(pkA, skB) == group_action(pkB, skA)
```

### Toy CSIDH

Implement a key exchange based on the supersingular curve below.

After having chosen the isogeny degrees $ℓ$, try to use both ideals $(π-1,ℓ)$ and $(π+1,ℓ)$ to define your random walks.

```python
p = 4*prod(prime_range(3,72)) - 1
E = EllipticCurve(GF(p), j=1728)
E
```

The curve is constructed so that 
$$E(𝔽_p) = p + 1 = 4 · \prod_{3 ≤ ℓ ≤ 72} ℓ$$
where the product runs over the primes. There are $19$ such primes, so we can define secrets as $19$-tuples of exponents in, say $[-2...2]$.

```python
len(prime_range(3,72))
```

```python
def secret_gen():
    from random import choices
    return choices(range(-2,3), k=19)
```

To evaluate the group action, we let positive exponents correspond to $(π-1, ℓ)$ and negative exponents to $(π+1, ℓ)$.

We use the method of taking a generator and multiplying by the cofactor. We also expoloit the fact that the points of $E[π+1]$ are defined over $𝔽_{p^2}$, however their $x$-coordinates are in $𝔽_p$, and thus they correspond to $𝔽_p$-rational points of the quadratic twist of $E$.

```python
def group_action(E, exp):
    ord = E.order()
    for (e, ell) in zip(exp, prime_range(3, 72)):
        if e < 0:
            E = E.quadratic_twist()
        for i in range(abs(e)):
            P, = E.gens()
            g = (ord // ell) * P
            E = E.isogeny(g).codomain()
        if e < 0:
            E = E.quadratic_twist()
    return E
```

And here we go!

```python
# Alice
skA = secret_gen()
pkA = group_action(E, skA)

# Bob
skB = secret_gen()
pkB = group_action(E, skB)

# Shared secret
%time group_action(pkA, skB).j_invariant() == group_action(pkB, skA).j_invariant()
```
